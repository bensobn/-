```python
from sympy import symbols, solve_univariate_inequality, Interval, oo ,sqrt, parse_expr, LessThan
import numpy as np
import random
from multiprocessing import Pool
from multiprocessing import cpu_count
```

```python

def problem_solve(a,x1,t1,v1):
  x = symbols('x')
  # 定义字符串表达式
  inequality_str = '{}+7*abs(x-{})+x+{}'.format(a[x1+7][t1+7], v1, v1)

  # 解析字符串表达式为 SymPy 表达式对象
  expression = parse_expr(inequality_str)

  # 创建 LessThan 对象
  inequality = LessThan(expression, 130)

  # 解不等式
  solution = solve_univariate_inequality(inequality, x)

  # 获取解的区间
  intervals = solution.args
  # 将区间表示为字符串
  interval_strs = []
  for interval in intervals:
      if isinstance(interval, Interval):
          word = str(interval.replace(oo, float('inf')).replace(-oo, float('-inf')).evalf())
          word = word.replace("Interval.open", "").replace("(", "[").replace(")", "]").replace("oo", "1000000")
          interval_strs.append(word)
      else:
          word = str(interval.as_set().replace(oo, float('inf')).replace(-oo, float('-inf')).evalf())
          word = word.replace("Interval.open", "").replace("(", "[").replace(")", "]").replace("oo", "1000000")
          interval_strs.append(word)

  # 移除字符串中的单引号并转换为列表形式的区间表示
  interval_lists = []
  for interval_str in interval_strs:
      interval_str = interval_str.replace("Interval[", "").replace("]", "")
      interval_list = [round(float(endpoint.strip()), 2) for endpoint in interval_str.split(",")]
      interval_lists.append(interval_list)

  if  len(interval_lists) == 0:
    interval_lists = 0
  else:  
    interval_lists = [max(interval_lists[0][0],0),interval_lists[1][1]]
  
  return interval_lists


```

```python
def count(tar):
    if  isinstance(tar,float) or isinstance(tar,int) :
        return tar
    else:
        
        value = tar[1]-tar[0]
        return value
```

```python
def calculate(a,b):
  
  if a == 0 or b == 0 :
    return 0
  else:
    result = [max(a[0], b[0]), min(a[1], b[1])]
    if result[1]-result[0]<0:
      return 0
    else:
      return result
```

```python
def get_shape(lst):
    if isinstance(lst, list):
        return [len(lst)] + get_shape(lst[0])
    else:
        return []
```

```python
def pool_sync(matrix,alcount,gettar,intent):
    global target_real
    global matrix_real
    pool_sz = cpu_count()
    with Pool(pool_sz) as p:
        p.starmap(mlt1, [(matrix,alcount,gettar,entint) for entint in intent] )
        p.close()
        p.join()
```
